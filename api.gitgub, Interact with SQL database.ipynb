## Go to https://api.github.com/repos/apache/hadoop/contributorsLinks to an external site. This is the Apache Hadoop Github Repo's contributors’ endpoint. Extract the JSON corresponding to the first 100 contributors from this API.

```python
import requests
import json
from bs4 import BeautifulSoup

# 100 per page is the maximum allowed by Github API, page_number is the variable
url = 'https://api.github.com/repos/apache/hadoop/contributors?per_page=100'
page = requests.get(url)
doc = BeautifulSoup(page.content, 'html.parser')

json_dict = json.loads(str(doc))

contributor_urls = []

# Get each of the contributor's URL
for contributor in json_dict:
    contributor_urls.append(contributor['url'])
    print(contributor['url'])
print(len(contributor_urls))
```

## For each of the 100 contributors extracted, write code that accesses their user information and extracts "login", "id", "location", "email", "hireable", "bio", "twitter_username", "public_repos", "public_gists", "followers", "following", "created_at" (and print those to screen)

```python

# Generated a github's PAT token as getting error of rate limits exceeded
token = ''
headers = {
    'Authorization': 'token ' + token,
    'User-Agent': 'Mozilla/5.0'
}


contributors_info = []
# Loop through each contributor's url
for contributor_url in contributor_urls:
    contributor = requests.get(contributor_url, headers = headers)
    contributor_data = contributor.json()
    
    info = {
            "login": contributor_data.get('login'),
            "id": contributor_data.get('id'),
            "location": contributor_data.get('location'),
            "email": contributor_data.get('email'),
            "hireable": contributor_data.get('hireable'),
            "bio": contributor_data.get('bio'),
            "twitter_username": contributor_data.get('twitter_username'),
            "public_repos": contributor_data.get('public_repos'),
            "public_gists": contributor_data.get('public_gists'),
            "followers": contributor_data.get('followers'),
            "following": contributor_data.get('following'),
            "created_at": contributor_data.get('created_at')
            }

    contributors_info.append(info)
    print(f"Collected information for {len(contributors_info)} contributor.")
    print(info)
    print("*"*50)
    
    
```

Based on above information of the contributors, login and id column both don't have any duplicates and both don't have null values, so both can be primary key in the sql table. I will be using "id" as the primary key. 

```python
import mysql.connector
from mysql.connector import Error
import warnings


import requests
import json
import codecs
from bs4 import BeautifulSoup

warnings.filterwarnings("ignore")
SQL_DB = "ucdavis"

# Method to create a table
def create_sql_table(SQL_TABLE, SQL_TABLE_DEF):
    try:
        #connect to server
        conn = mysql.connector.connect(host='localhost',
                                       user='root',
                                       password='') # password deleted
        cursor = conn.cursor()
        
        query = "CREATE DATABASE IF NOT EXISTS " + SQL_DB
        print(query)
        cursor.execute(query);
        
        query = "CREATE TABLE IF NOT EXISTS " + SQL_DB + "." + SQL_TABLE + " " + SQL_TABLE_DEF + ";";
        print(query)
        cursor.execute(query);
        cursor.close()
        conn.close()
        return

    except IOError as e:
        print(e)
```

```python

SQL_TABLE = "contributors"
SQL_TABLE_DEF = """
(
    id INT PRIMARY KEY, 
    login VARCHAR(255) ,
    location VARCHAR(255),
    email VARCHAR(255),
    hireable INT,
    bio TEXT,
    twitter_username VARCHAR(255),
    public_repos INT,
    public_gists INT,
    followers INT,
    following INT,
    created_at DATETIME
)
"""
from datetime import datetime
def insert_contributors_data(contributors_info):
    try:
        # Connect to the MySQL server
        conn = mysql.connector.connect(host='localhost', user='root', password='', database=SQL_DB) # password deleted
        cursor = conn.cursor()

        # SQL query to insert data
        add_contributor = (f"INSERT INTO {SQL_TABLE} "
                           "(id, login, location, email, hireable, bio, twitter_username, public_repos, "
                           "public_gists, followers, following, created_at) "
                           "VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)")

        # Iterate through each contributor's data
        for contributor in contributors_info:
            # Convert the created_at to the MySQL datetime format
            created_at = contributor.get('created_at')
            if created_at:
                created_at = datetime.strptime(created_at, '%Y-%m-%dT%H:%M:%SZ').strftime('%Y-%m-%d %H:%M:%S')
            else:
                created_at = None 
            
            # Lable hireable
            hireable = contributor.get('hireable')
            
            if hireable is not None:
                hireable = int(hireable)
            else:
                hireable = 0 
            # Get the contributor's data in the same order as the columns in the INSERT statement
            contributor_data = (
                contributor.get('id'), 
                contributor.get('login'), 
                contributor.get('location'), 
                contributor.get('email'), 
                hireable,  
                contributor.get('bio'), 
                contributor.get('twitter_username'), 
                contributor.get('public_repos'), 
                contributor.get('public_gists'), 
                contributor.get('followers'), 
                contributor.get('following'), 
                created_at
            )
            # Insert data
            print(contributor_data)
            cursor.execute(add_contributor, contributor_data)
        
        # Commit the changes
        conn.commit()

    except Error as e:
        print(e)

    finally:
        cursor.close()
        conn.close()
```

```python
create_sql_table(SQL_TABLE, SQL_TABLE_DEF)
insert_contributors_data(contributors_info)
```
